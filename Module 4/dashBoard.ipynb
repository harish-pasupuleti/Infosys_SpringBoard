```python
import pandas as pd
import dash
from dash import dcc, html
import plotly.express as px
from pymongo import MongoClient
from dash.dependencies import Input, Output

# ✅ Step 1: Connect to MongoDB and Fetch Data
client = MongoClient("mongodb+srv://harishpasupuleti18:QzPqXVRXmYAYygHZ@cluster0.knp4m.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0")
db = client["HotelAnalytics"]

# Load data from MongoDB
booking_df = pd.DataFrame(list(db["BookingData"].find()))
dining_df = pd.DataFrame(list(db["DiningInfo"].find()))
reviews_df = pd.DataFrame(list(db["ReviewsData"].find()))

# ✅ Step 2: Data Preprocessing

# Convert date columns
booking_df["check_in_date"] = pd.to_datetime(booking_df["check_in_date"], format="%d-%m-%Y", errors="coerce")
booking_df["check_out_date"] = pd.to_datetime(booking_df["check_out_date"], format="%d-%m-%Y", errors="coerce")

# Calculate the length of stay
booking_df["length_of_stay"] = (booking_df["check_out_date"] - booking_df["check_in_date"]).dt.days

# Convert dining dates
dining_df["check_in_date"] = pd.to_datetime(dining_df["check_in_date"], errors="coerce")

# Convert review ratings to numeric
reviews_df["Rating"] = pd.to_numeric(reviews_df["Rating"], errors="coerce")

# ✅ Step 3: Initialize Dash App
app = dash.Dash(__name__)

# ✅ Step 4: Define Layout
app.layout = html.Div(children=[
    html.H1("Hotel Analytics Dashboard", style={"text-align": "center"}),

    # 🔹 Dropdown to Select Dashboard
    dcc.Tabs(id="dashboard-tabs", value="bookings", children=[
        dcc.Tab(label="Hotel Booking Insights", value="bookings"),
        dcc.Tab(label="Dining Insights", value="dining"),
        dcc.Tab(label="Reviews Analysis", value="reviews"),
    ]),

    html.Div(id="dashboard-content")
])

# ✅ Step 5: Callbacks to Render Different Dashboards
@app.callback(
    Output("dashboard-content", "children"),
    [Input("dashboard-tabs", "value")]
)
def update_dashboard(selected_tab):
    if selected_tab == "bookings":
        # 🔹 Line Chart: Bookings Over Time
        fig1 = px.line(booking_df, x="check_in_date", title="Bookings Trend Over Time")

        # 🔹 Pie Chart: Preferred Cuisine Analysis
        fig2 = px.pie(dining_df, names="Preferred Cusine", title="Preferred Cuisine Analysis")

        # 🔹 Bar Chart: Average Length of Stay (Fixed)
        avg_stay_df = booking_df.groupby(booking_df["check_in_date"].dt.to_period("M"))["length_of_stay"].mean().reset_index()
        avg_stay_df["check_in_date"] = avg_stay_df["check_in_date"].astype(str)  # Convert to string

        fig3 = px.bar(avg_stay_df, x="check_in_date", y="length_of_stay", title="Average Length of Stay (Monthly)")

        return html.Div([
            dcc.Graph(figure=fig1),
            dcc.Graph(figure=fig2),
            dcc.Graph(figure=fig3),
        ])

    elif selected_tab == "dining":
        # 🔹 Pie Chart: Average Dining Cost by Cuisine
        fig1 = px.pie(dining_df, names="Preferred Cusine", values="price_for_1", title="Average Dining Cost by Cuisine")

        # 🔹 Line Chart: Customer Count Over Time (Fixed)
        customer_count_df = dining_df.groupby(dining_df["check_in_date"].dt.to_period("M")).size().reset_index(name="count")
        customer_count_df["check_in_date"] = customer_count_df["check_in_date"].astype(str)  # Convert to string

        fig2 = px.line(customer_count_df, x="check_in_date", y="count", title="Customer Count Over Time")

        return html.Div([
            dcc.Graph(figure=fig1),
            dcc.Graph(figure=fig2),
        ])

    elif selected_tab == "reviews":
        # 🔹 Pie Chart: Sentiment Analysis (Positive, Neutral, Negative Reviews)
        sentiment_counts = reviews_df["Rating"].apply(lambda x: "Positive" if x >= 7 else "Neutral" if x >= 4 else "Negative").value_counts()
        fig1 = px.pie(names=sentiment_counts.index, values=sentiment_counts.values, title="Sentiment Analysis")

        # 🔹 Histogram: Rating Distribution
        fig2 = px.histogram(reviews_df, x="Rating", title="Rating Distribution", nbins=10)

        return html.Div([
            dcc.Graph(figure=fig1),
            dcc.Graph(figure=fig2),
        ])

# ✅ Step 6: Run the Dash App
if __name__ == "__main__":
    app.run_server(debug=True, port=8060)

```
